```python
# Missing salary values are filled with the median because only placed students have salary values.
df["salary"] = df["salary"].fillna(df["salary"].median())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = df[col].str.strip().str.title()
```

```python

df.rename(columns={
    "status":"placement_status",
    "salary":"salary_inr"
}, inplace=True)
```

```python
df["salary_inr"] = df["salary_inr"].astype(float)
```

```python
df["Salary_Lakh"] = df["salary_inr"] / 100000
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
